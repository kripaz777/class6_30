```
output_data = []
tags = soup.table
tags = tags.find_all('tr')
tags = tags[1:-3]
n = len(tags)
for i in range(n):
  output_data.append(tags[i].get_text())


print(output_data)
```

```
final_data = []
n = len(output_data)

for i in range(n):
  x = output_data[i].split('\n')
  y = x[1:-1]
  final_data.append(y)

final_data
```

```
import csv
n = len(final_data)
with open('stock.csv','a') as csvFile:
  for i in range(n):
    w = csv.writer(csvFile)
    w.writerow(final_data[i])
```
